## Split dataset

This code splits the original dataset into `train` and `val` subsets.

```python
import random
import shutil
from pathlib import Path

dataset_dir = Path("animals141/dataset/dataset")
output_dir = Path("data")
val_ratio = 0.2
seed = 42


random.seed(seed)

train_dir = output_dir / "train"
val_dir   = output_dir / "val"

# Create output directories
train_dir.mkdir(parents=True, exist_ok=True)
val_dir.mkdir(parents=True, exist_ok=True)

classes = [d.name for d in dataset_dir.iterdir() if d.is_dir()]
print(f"Osztályok: {classes}")

# Iterate through each class and split images
for cls in classes:
    src_dir = dataset_dir / cls
    imgs = list(src_dir.glob("*.*"))
    random.shuffle(imgs)

    # Split images into train and val
    n_val = int(len(imgs) * val_ratio)
    val_imgs = imgs[:n_val]
    train_imgs = imgs[n_val:]

    # Copy images to respective directories
    for split, split_imgs in zip(["train", "val"], [train_imgs, val_imgs]):
        split_cls_dir = output_dir / split / cls
        split_cls_dir.mkdir(parents=True, exist_ok=True)
        for img_path in split_imgs:
            dst_path = split_cls_dir / img_path.name
            shutil.copy2(img_path, dst_path)
```
